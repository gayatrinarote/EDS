```python
#Arithematic
import numpy as np
array3 = np.loadtxt("/FirstDecade(1).csv",delimiter=',',dtype=str,skiprows=1)
print(array3)
sal=[]
exp=[]
for i in array3:
  sal.append((i[3]))
  exp.append((i[4]))
  print(sal)
  print(exp)

  #converting list to array
  arr_sal=np.array(sal)
  arr_exp=np.array(exp)

  #displaying array
  print("Array1:",arr_sal)
  print("Array2:",arr_exp)
```
